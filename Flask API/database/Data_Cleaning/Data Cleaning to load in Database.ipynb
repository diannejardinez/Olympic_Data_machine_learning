```python
import pandas as pd
```

### Importing 'athlete_events.csv'

```python
athlete_events = pd.read_csv('athlete_events.csv')
```

```python
athlete_events.head()
```

### Importing 'noc_regions.csv'

```python
noc_regions = pd.read_csv('noc_regions.csv')
```

```python
noc_regions.head()
```

### Combining the two dataframes

```python
combined_df = pd.merge(athlete_events, noc_regions, on="NOC", how="left")
```

```python
combined_df.head()
```

```python
combined_df.loc[combined_df["Country"].isnull(), ["NOC","Team","Country"]].drop_duplicates()
```

```python
combined_df.loc[combined_df["NOC"]=="SGP", "Country"] = "Singapore"
combined_df.loc[combined_df["NOC"]=="ROT", "Country"] = "Refugee Olympic Athletes"
combined_df.loc[combined_df["NOC"]=="UNK", "Country"] = "Unknown"
combined_df.loc[combined_df["NOC"]=="TUV", "Country"] = "Tuvalu"
```

```python
olympics_df = combined_df[["ID", "Name", "Sex", "Age", "Height", "Weight", "Country",\
                          "NOC", "Games", "Year", "Season", "City", "Sport", "Event","Medal"]].copy()
```

```python
olympics_df.head()
```
